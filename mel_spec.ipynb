```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
scale_file="D:\Music samples\iveen-harps-sample-pack\SD_IVEEN_Harps_SamplePack\Harp One-Shots\SD_IVEENHarps_oneshot_01_C.wav"
```

```python
ipd.Audio(scale_file)
```

```python
scale,sr=librosa.load(scale_file) #loading features in librosa
```

```python
filter_banks=librosa.filters.mel(n_fft=2048,sr=22050,n_mels=10) #framesize,samplerate and no of bands
```

```python
filter_banks.shape #shape of the graph we ultimately gonna plot
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(filter_banks,
                        sr=sr,
                        x_axis='linear')
plt.colorbar(format='%+2.f')
plt.show()
```

```python
mel_spectogram=librosa.feature.melspectrogram(scale,sr=sr,n_fft=2048,hop_length=512,n_mels=10)
```

```python
mel_spectogram.shape 
```

```python
log_mel_spec=librosa.power_to_db(mel_spectogram)
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(log_mel_spec,
                        x_axis='time',
                        y_axis='mel',
                        sr=sr)
plt.colorbar(format='%+2.f')
plt.show()
```
